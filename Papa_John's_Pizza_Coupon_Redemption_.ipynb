```python
A = df.drop(columns=["Code", "Description", "REDEEMED"])
B = df["REDEEMED"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(A, B, test_size=0.2, random_state=7)
```

```python
X_train.shape
```

```python
model = Sequential()
```

```python
model.add(Dense(units=36, activation='relu', input_dim=4))
model.add(Dense(units=1, activation='linear'))
```

```python
model.compile(optimizer=Adam(), loss='mean_squared_error')
```

```python
model.fit(X_train, y_train, batch_size=5, epochs=124)
```

```python
y_predict = model.predict(X_test)
```
